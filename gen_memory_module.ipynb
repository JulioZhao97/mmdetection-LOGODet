```python
# gen index between img and style
import os
import pickle
from xml.etree import ElementTree as ET

style_index = {}
anno_all = os.listdir('/data/zhaozhiyuan/tb_variation/VOCdevkit_all/VOC2007/Annotations')
anno_all = [anno.split('.')[0] for anno in anno_all]
for anno in anno_all:
    pic_file = '/data/zhaozhiyuan/tb_variation/VOCdevkit_all/VOC2007/JPEGImages/{}.jpg'.format(anno)
    anno_file = '/data/zhaozhiyuan/tb_variation/VOCdevkit_all/VOC2007/Annotations/{}.xml'.format(anno)
    anno = ET.parse(anno_file).getroot()
    style = anno.find('object').find('style').text
    style_index[pic_file] = int(style)
f = open('style_index.pt', 'wb')
pickle.dump(style_index, f)
f.close()
```
